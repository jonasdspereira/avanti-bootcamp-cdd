# Perguntas de Partida e Hipóteses

Nesta seção, levantamos questões que podem ser exploradas para entender melhor os dados de diamantes.

### Perguntas de Partida
1. Existe uma relação direta entre o preço e o quilate do diamante?
2. Como as variáveis de cor, clareza e corte influenciam no preço do diamante?
3. Existe algum padrão geográfico de distribuição de preços?

### Hipóteses

- **Hipótese 2**: Diamantes com corte de melhor qualidade possuem valores superiores.
- **Hipótese 3**: Diamantes com coloração mais próxima do incolor têm preços mais altos, independentemente do quilate.


## 3. Insights

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Verificar a relação entre preço e quilate
plt.figure(figsize=(10, 6))
sns.scatterplot(x='carat', y='price', data=df, alpha=0.3)
plt.title('Relação entre Quilate e Preço')
plt.xlabel('Quilate')
plt.ylabel('Preço')
plt.show()

# Análise de como o corte afeta o preço
plt.figure(figsize=(10, 6))
sns.boxplot(x='cut', y='price', data=df)
plt.title('Preço por Qualidade de Corte')
plt.xlabel('Qualidade do Corte')
plt.ylabel('Preço')
plt.show()

# Análise da influência da cor no preço
plt.figure(figsize=(10, 6))
sns.boxplot(x='color', y='price', data=df)
plt.title('Preço por Cor do Diamante')
plt.xlabel('Cor')
plt.ylabel('Preço')
plt.show()

# Análise de influência da clareza no preço
plt.figure(figsize=(10, 6))
sns.boxplot(x='clarity', y='price', data=df)
plt.title('Preço por Clareza do Diamante')
plt.xlabel('Clareza')
plt.ylabel('Preço')
plt.show()
```
